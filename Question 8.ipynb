```
import cv2
import matplotlib.pyplot as plt
import math
import numpy as np
import scipy.fftpack as fp
from google.colab import files
upload = files.upload()

img = cv2.imread('fig3.tif',0)
(P,Q) = img.shape
```

```
def HighpassGaussianFilter(P,Q):
    H = np.zeros((P,Q), dtype=complex)
    p = int(P/2)
    q = int(Q/2)
    for u in range(-p, p):
        for v in range(-q,q):
            H[u+p, v+q] = 1 - math.exp(-1* math.sqrt((u)**2 + (v)**2)/(2*900))

    return H

def performFunction(img,H):
    plt.imshow(img,cmap='gray')
    plt.show()
    F1 = fp.fft2((img).astype(float))
    F2 = fp.fftshift(F1)
    F2= F2*H
    return F2

H = HighpassGaussianFilter(P,Q)

for i in range(100):
    print(i)
   
    F2 = performFunction(img,H)
    img = fp.ifft2(fp.ifftshift(F2)).astype(float)

```
