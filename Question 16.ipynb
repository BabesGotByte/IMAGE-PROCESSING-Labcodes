```
img=cv2.imread('Image2.tif',0)
plt.imshow(img)
```

```
#Fourier transform
f_img=np.fft.fft2(img)
fshift2=np.fft.fftshift(f_img)
mag_spec=20*np.log(np.abs(fshift2))
plt.imshow(mag_spec)
```

```
#Filter
def filter(m,n):
  uu=np.ones((m,n))
  uu[:,n//2-5:n//2+5]=0
  uu[m//2-5:m//2+5,n//2-5:n//2+5]=1
  return uu
```

```
HH=filter(img.shape[0],img.shape[1])
GG=HH*fshift2
```

```
#Inverse Fourier transform
fd2_ishift = np.fft.ifftshift(GG)
imgd2_back = np.fft.ifft2(fd2_ishift)
imgd2_back = np.abs(imgd2_back)
plt.imshow(imgd2_back)
```
