```
import cv2
from math import log
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('cameraman.tif',0)

h, w = img.shape[:2]

for y in range(0, h):
  for x in range(0, w):
    if(img[x,y]<5 or img[x,y]>50):
      img[x,y] = 255
    else:
      img[x,y] = img[x,y]
       	 
       	 
cv2_imshow(img)
cv2.waitKey(0)
```
