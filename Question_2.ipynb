```
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt

BLUE = [255,0,0]
img1 = cv2.imread('lenna.png')
img2 = cv2.imread('lenna.png')

#mirror image
myimg = img1[::1,::-1]

#horizontally concatenate
h1, w1 = img1.shape[:2]
h2, w2 = img2.shape[:2]

vis1 = np.zeros((max(h1, h2), w1+w2,3), np.uint8)
vis1[:h1, :w1,:3] = img1
vis1[:h2, w1:w1+w2,:3] = img2


#vertically concatenate
h1, w1 = img1.shape[:2]
h2, w2 = img2.shape[:2]

vis2 = np.zeros(((h1+h2), max(w1,w2),3), np.uint8)
vis2[:h1, :w1,:3] = img1
vis2[h1:h1+h2, :w2,:3] = img2


visible1 = cv2.copyMakeBorder(vis1,50,50,50,50,cv2.BORDER_CONSTANT,value=BLUE)
visible2 = cv2.copyMakeBorder(vis2,50,50,50,50,cv2.BORDER_CONSTANT,value=BLUE)
visible3 = cv2.copyMakeBorder(img1,50,50,50,50,cv2.BORDER_CONSTANT,value=BLUE)
visible4 = cv2.copyMakeBorder(myimg,50,50,50,50,cv2.BORDER_CONSTANT,value=BLUE)

plt.subplot(231),plt.imshow(visible3,'gray'),plt.title('ORI IMA')
plt.subplot(232),plt.imshow(visible4,'gray'),plt.title('MIRROR IMA')
plt.subplot(234),plt.imshow(visible1,'gray'),plt.title('HORI CONC')
plt.subplot(235),plt.imshow(visible2,'gray'),plt.title('VERT CONC')

plt.show()
```
