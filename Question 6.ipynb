```
import cv2 as cv2
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt

image = cv2.imread('pout.tif').astype(np.float32) / 255
plt.imshow(image)

print("mean: ")
print(image.mean())
print("variance: ")
print(image.std()*image.std())
```

```
image -= image.mean()
plt.imshow(image)
```

```
image /= image.std()
plt.imshow(image)
```

```
image /= image.std()*image.std()
plt.imshow(image)
```
